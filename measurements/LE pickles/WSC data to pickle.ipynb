```python
import pandas as pd
from IPython.core.display import display
```

```python
STATIONS = pd.read_csv('STATIONS.csv', index_col = 0);
DLY_LEVELS = pd.read_csv('DLY_LEVELS.csv')
DLY_FLOWS = pd.read_csv('DLY_FLOWS.csv')

print("    Stations in the HYDAT database = ", len(STATIONS.index))
```

```python
# bounding region
lngW = -85.
lngE = -75.
latS = 41.
latN = 44.
```

```python
# locate monitoring stations within bounding region
stationList = STATIONS[(STATIONS['LATITUDE'] >= latS) & \
         (STATIONS['LATITUDE'] <= latN) & \
         (STATIONS['LONGITUDE'] <= lngE) & \
         (STATIONS['LONGITUDE'] >= lngW)].STATION_NUMBER.values
```

```python
# reduce data sets to the area of interest
WSC_STATIONS = STATIONS[STATIONS.STATION_NUMBER.isin(stationList)]
DLY_LEVELS = DLY_LEVELS[DLY_LEVELS.STATION_NUMBER.isin(stationList)]
DLY_FLOWS = DLY_FLOWS[DLY_FLOWS.STATION_NUMBER.isin(stationList)]
```

```python
DLY_FLOWS = DLY_FLOWS[DLY_FLOWS.YEAR>1995]
```

```python
# add columns to STATIONS for level and flow stations
WSC_STATIONS['Level'] = ''
WSC_STATIONS['Flow'] = ''
```

```python
# mark level and flow stations
WSC_STATIONS.loc[WSC_STATIONS.STATION_NUMBER.isin(DLY_LEVELS.STATION_NUMBER), "Level"] = True
WSC_STATIONS.loc[WSC_STATIONS.STATION_NUMBER.isin(DLY_FLOWS.STATION_NUMBER), "Flow"] = True


```

```python
print("Stations within region of interest = ", len(WSC_STATIONS.index))
print("DLY_FLOWS records = ", len(DLY_FLOWS.index))
print("DLY_LEVELS records = ", len(DLY_LEVELS.index))
display(WSC_STATIONS.ix[:,['Level','Flow','STATION_NAME','LATITUDE','LONGITUDE',]])

WSC_STATIONS.to_pickle('data/WSC_STATIONS')
```

```python
def getFlowsWSC(s):
#     s = [s for s in WSC_STATIONS[WSC_STATIONS['Flow']==True].STATION_NUMBER][0]
    ts = {}
    data = DLY_FLOWS[DLY_FLOWS['STATION_NUMBER'] == s]
    for k in data.index:
            mo = str(data.ix[k,'MONTH'])
            yr = str(data.ix[k,'YEAR'])
            for n in range(1,data.ix[k,'NO_DAYS']+1):
#                 print(data.ix[k])
#                 print(float(data.ix[k,'FLOW'+str(n)]))
                ts[pd.to_datetime(yr+'-'+mo+'-'+str(n))] = data.ix[k,'FLOW'+str(n)]
                
    ts = pd.Series(ts)
    ts.name = str(s) +': ' + STATIONS[STATIONS['STATION_NUMBER']==s].STATION_NAME.values[0] + \
            ' from ' + '{0}'.format(ts.index[0].year) + \
            ' to ' + '{0}'.format(ts.index[-1].year)
#     #drop initial and terminal null entries
#     j = 0
#     while j<len(ts.index) and pd.isnull(ts.ix[j]):
#         j += 1
#     k = len(ts.index) - 1
#     while k>=j and pd.isnull(ts.ix[k]):
#         k += -1
#     return ts[j:k+1]
    return ts

WSC_FLOWS = pd.DataFrame({s: getFlowsWSC(s) for s in WSC_STATIONS[WSC_STATIONS['Flow']==True].STATION_NUMBER})
WSC_FLOWS.to_pickle('data/WSC_FLOWS')
WSC_FLOWS.tail()
```

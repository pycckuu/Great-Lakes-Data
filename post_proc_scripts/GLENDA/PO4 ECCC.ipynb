```python
df2=df[pd.to_numeric(df['DEPTH_FROM'], errors='coerce').notnull()]
df2=df2[pd.to_numeric(df2['VALUE'], errors='coerce').notnull()]
df2['DEPTH_FROM']=df2[['DEPTH_FROM']].apply(pd.to_numeric)
df2['DEPTH_FROM']=df2[['DEPTH_FROM']]*-1
df2['VALUE']=df2[['VALUE']].apply(pd.to_numeric)
df2['Year'] = df.STN_DATE.dt.year
df2['Month'] = df.STN_DATE.dt.month
df2['Day'] = df.STN_DATE.dt.day
df2['DOY'] = df.STN_DATE.dt.dayofyear
df2['WOY'] = df.STN_DATE.dt.weekofyear
```

```python
a = df2[df2['PSN']==879]['STN_DATE'].unique()
```

```python
np.sort(a)
```

```python
df2[(df2['PSN']==879)].groupby(['STN_DATE', 'DEPTH_FROM']).mean()['VALUE'].unstack(level=0)
```

```python
df2[(df2['PSN']==879) & (df2.ABBREV == 'TP')].groupby(['STN_DATE', 'DEPTH_FROM']).mean()['VALUE'].unstack(level=0).to_csv('oxy_ecc_all_dates.csv')
```

```python
df2[(df2['PSN']==879) & (df2.ABBREV == 'TP')].groupby(['STN_DATE', 'DEPTH_FROM']).mean()['VALUE'].unstack(level=0).dropna(axis=1,how='all')
```

```python
df2[(df.PSN.isin(ee_ECCC_stations)) & (df2.ABBREV == 'SR P')].groupby(['Year','Month', 'DEPTH_FROM']).mean()['VALUE'].unstack(level=0).unstack(level=0).dropna(axis=1,how='all').to_csv('SRP_eccc_ee.csv')
```

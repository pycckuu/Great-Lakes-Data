```python
years=np.linspace(2000, 2015, 16)
months = df.MONTH.unique()[:2]
dfs = {}
for y in years:
    for m in months:
        dfs[" ".join([str(y),str(m)]).strip()]=df[(df['YEAR']==y) & (df['MONTH']==m)& (df['STATION_ID']=='ER15M')]
```

```python
df2=df[pd.to_numeric(df['SAMPLE_DEPTH_M'], errors='coerce').notnull()]
df2=df2[pd.to_numeric(df2['VALUE_1'], errors='coerce').notnull()]
df2['SAMPLE_DEPTH_M']=df2[['SAMPLE_DEPTH_M']].apply(pd.to_numeric)
df2['SAMPLE_DEPTH_M']=df2[['SAMPLE_DEPTH_M']]*-1
df2['VALUE_1']=df2[['VALUE_1']].apply(pd.to_numeric)
```

```python
df2[df2['STATION_ID']=='ER15M'][['SAMPLING_DATE', 'SAMPLE_DEPTH_M','VALUE_1']].sort_values(by=['SAMPLE_DEPTH_M']).to_csv('Temperature_ER15M.csv', index=False)
```

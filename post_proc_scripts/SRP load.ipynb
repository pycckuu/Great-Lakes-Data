```python
import pandas as pd
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('/Users/imarkelo/git/Great-Lakes-Data/measurements/ncwqr.org/Sandusky_River_data.xlsx')
```

```python
data['load mg/s'] = pd.to_numeric(data['Flow, CFS'], errors='coerce')*0.0283168*data['SRP, mg/L, as P']/1000
```

```python
data['load mg/s'] = data['load mg/s']*(data['load mg/s']>0)
```

```python
data['cum']=data['Days since 741001']*data['load mg/s']
```

```python
data['cum'] = data['Days since 741001'].diff()*data['load mg/s']
```

```python
plt.plot(data['Datetime (date and time of sample collection)'], data['cum'].cumsum())
```

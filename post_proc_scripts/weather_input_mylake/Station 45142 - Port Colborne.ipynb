```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 32})
sns.set_context("poster", rc={"font.size":34,"axes.titlesize":34,"axes.labelsize":34, "lines.linewidth": 2})  
plt.rcParams['figure.figsize'] = 18, 12

pd.options.display.max_columns = 999
pd.options.display.max_rows = 9999


from datetime import datetime
str(datetime.now())


```

```python
df = pd.read_csv('../../measurements/Custom/c45142.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['DATE'] = df.apply(lambda x:datetime.strptime("{0}".format(x['DATE']), "%m/%d/%Y %H:%M"),axis=1)
```

```python
df.tail()
```

```python
df.set_index('DATE', inplace=True)
```

```python
df.tail()
```

```python
for c in df.columns[1:]:
    df[c] = pd.to_numeric(df[c], errors='coerce')
    df[c] = df[c].interpolate(method='linear', limit=30)
```

```python
df.tail()
```

```python
x2= pd.date_range('1983-04-01','2015-12-31',freq='D')
```

```python
df2 = df.reindex(x2)
```

```python
df2.describe()
```

```python
daily = pd.DataFrame()
for c in df.columns[1:]:
    daily[c] = df[c].resample('D').mean()
#     daily[c] = daily[c].interpolate(method='linear', limit=100)
```

```python
daily.tail()
```

```python
daily2 = daily.reindex(x2)
daily2.tail()
```

```python
daily2.to_csv('Station 45142.csv')
```

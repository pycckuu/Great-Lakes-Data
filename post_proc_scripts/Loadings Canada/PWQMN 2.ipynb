```python
dfs = [pwqmn_rawdata_2016, pwqmn_rawdata_2015, pwqmn_rawdata_2014, 
       pwqmn_rawdata_2013, pwqmn_rawdata_2012, pwqmn_rawdata_2011, 
       pwqmn_rawdata_2010, pwqmn_rawdata_2009, pwqmn_rawdata_2008, 
       pwqmn_rawdata_2007, pwqmn_rawdata_2006, pwqmn_rawdata_2005,
      pwqmn_rawdata_2004, pwqmn_rawdata_2003, pwqmn_rawdata_2002,
      pwqmn_rawdata_2001, pwqmn_rawdata_2000, pwqmn_rawdata_1996_1999]

```

```python
dfsn = []
for df in dfs:
    df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
    for c in df.columns:
        df = df.rename(index=str, columns={c:c.replace("_", " ")})
    df.rename({"STATION NO":"STATION", "LIMS_R": 'LIMS RCODE', "FIELD_NO":"FIELD NO", "FIELDNO":"FIELD NO"}, inplace=True)
    dfsn.append(df)
```

```python
pwqmn_rawdata = pd.concat(dfsn, ignore_index=True)
```

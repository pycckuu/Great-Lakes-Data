```python
import pandas as pd
```

```python
glni = pd.read_excel('../../measurements/GLNI/Canadian_Tributaries_concentrations_Serghei.xlsx', sheet_name='Data_GLNI_Erie tribs')
```

```python
glni.columns
```

```python
species = ['Phosphorus total', 'Phosphorus total filtered', 'Phosphorus soluble reactive', 'Phosphorus particulate', 'Phosphorus total dissolved',
           'Nitrogen total', 'Nitrogen Kjeldahl total', 'Nitrogen particulate', 'Nitrite and nitrate filtered', 'Ammonia filtered', 'Nitrogen total dissolved',
           'SiO2', 'Silica soluble reactive']
```

```python
for river in glni['SiteName'].unique():
    df = glni[glni['SiteName'] == river]
    df['SampleDateTime'] = pd.to_datetime(df['SampleDateTime']).dt.date
    elements = []

    for elem in sorted(df['VariableCommonName'].unique()):
        print(elem)
        df_1 = df[(df['VariableCommonName'] == elem) & (df['SampleQaTypeCode'] != 'BB') & (df['SampleQaTypeCode'] != 'MB')][['SampleDateTime','Value', 'Unit']]
        if len(df_1['Unit'].unique())!=1:
            print('fail')
            break
        df_2 = df_1.groupby('SampleDateTime').mean().reset_index()[['SampleDateTime','Value']]
        df_2['Unit'] = df_1['Unit'].unique()[0]
        df_2.rename(columns={'Unit':'Unit '+elem, 'Value': elem}, index=str, inplace=True)
        elements.append(df_2)

    res = elements[0]
    for d in elements[1:]:
        res = pd.merge(res, d, on='SampleDateTime', how='outer')

    res.to_csv("GLNI_"+river.lower().replace(" ", "_")+"_"+df['SiteCode'].unique()[0]+'.csv')
```

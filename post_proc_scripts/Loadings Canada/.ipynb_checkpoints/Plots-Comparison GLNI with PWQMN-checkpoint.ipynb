```python
csv_glni = './eastern_basin/nanticoke_creek/GLNI/GLNI_nanticoke_creek_at_nanticoke_ON02GC1000.csv'

df = pd.read_csv(csv_glni)
df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
fig, ax = plt.subplots()
sns.scatterplot(x='SampleDateTime', y='Phosphorus total', data=df)

csv_PWQMN = './eastern_basin/nanticoke_creek/PWQMN/PWQMN_nanticoke_creek_16016400102.csv'
df = pd.read_csv(csv_PWQMN)
df['DATE'] = pd.to_datetime(df.DATE)
sns.scatterplot(x='DATE', y='PHOSPHORUS,UNFILTERED TOTAL', data=df)

ax.grid()
# sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
ax.set_ylabel('Phosphorus total, mg/L')
ax.set_xlabel('')
# assign locator and formatter for the xaxis ticks.
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.legend(['GLNI ON02GC1000', 'PWQMN 16016400102'])
ax.set_title('Phosphorus total, mg/L')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.savefig('./eastern_basin/nanticoke_creek/nanticoke_creek_TP.pdf')
```

```python
pairs = [['Chloride dissolved', 'CHLORIDE,         UNFIL.REAC'],
         ['Nitrogen total','NITROGEN; TOTAL'],
         ['Phosphorus total', 'PHOSPHORUS,UNFILTERED TOTAL'],
         ['Nitrogen Kjeldahl total','NITROGEN,TOT,KJELDAHL/UNF.REA']
        ]
```

```python
csv_glni = './eastern_basin/nanticoke_creek/GLNI/GLNI_nanticoke_creek_at_nanticoke_ON02GC1000.csv'
csv_PWQMN = './eastern_basin/nanticoke_creek/PWQMN/PWQMN_nanticoke_creek_16016400102.csv'

for p in pairs:
    df = pd.read_csv(csv_glni)
    df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
    df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
    fig, ax = plt.subplots()
    sns.scatterplot(x='SampleDateTime', y=p[0], data=df)


    df = pd.read_csv(csv_PWQMN)
    df['DATE'] = pd.to_datetime(df.DATE)
    sns.scatterplot(x='DATE', y=p[1], data=df)

    ax.grid()
    # sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
    ax.set_ylabel(p[0]+', mg/L')
    ax.set_xlabel('')
    # assign locator and formatter for the xaxis ticks.
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.legend(['GLNI '+csv_glni.split("_")[-1][:-3], 'PWQMN '+csv_PWQMN.split("_")[-1][:-4]])
    ax.set_title(p[0]+', mg/L')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.savefig(("/").join(csv_glni.split("/")[:3])+"/"+csv_glni.split("/")[2]+"_"+p[0].lower().replace(" ","_")+'.pdf')
```

```python
csv_glni = './eastern_basin/grand_river/GLNI/GLNI_grand_river_at_york_ON02GC1600.csv'
csv_PWQMN = './eastern_basin/grand_river/PWQMN/PWQMN_grand_river_16018403502.csv'

for p in pairs:
    df = pd.read_csv(csv_glni)
    df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
    df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
    fig, ax = plt.subplots()
    sns.scatterplot(x='SampleDateTime', y=p[0], data=df)


    df = pd.read_csv(csv_PWQMN)
    df['DATE'] = pd.to_datetime(df.DATE)
    sns.scatterplot(x='DATE', y=p[1], data=df)

    ax.grid()
    # sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
    ax.set_ylabel(p[0]+', mg/L')
    ax.set_xlabel('')
    # assign locator and formatter for the xaxis ticks.
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax.legend(['GLNI '+csv_glni.split("_")[-1][:-4], 'PWQMN '+csv_PWQMN.split("_")[-1][:-4]])
    ax.set_title(p[0]+', mg/L')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.savefig(("/").join(csv_glni.split("/")[:3])+"/"+csv_glni.split("/")[2]+"_"+p[0].lower().replace(" ","_")+'.pdf')
```

```python
csv_glni = './eastern_basin/big_creek/GLNI/GLNI_big_creek_near_walsingham_ON02GC1200.csv'

csv_PWQMN_1 = './eastern_basin/big_creek/PWQMN/PWQMN_big_creek_16012400302.csv'
csv_PWQMN_2 = './eastern_basin/big_creek/PWQMN/PWQMN_big_creek_16012401002.csv'
csv_PWQMN_3 = './eastern_basin/big_creek/PWQMN/PWQMN_big_creek_16012401102.csv'


for p in pairs:
    df = pd.read_csv(csv_glni)
    df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
    df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
    fig, ax = plt.subplots()
    sns.scatterplot(x='SampleDateTime', y=p[0], data=df)
    
    leg = ['GLNI '+csv_glni.split("_")[-1][:-4]]
    for csv_PWQMN in [csv_PWQMN_1, csv_PWQMN_2, csv_PWQMN_3]:
        df = pd.read_csv(csv_PWQMN)
        df['DATE'] = pd.to_datetime(df.DATE)
        if df[~df[p[1]].isnull()].shape[0]>0:
            sns.scatterplot(x='DATE', y=p[1], data=df)
            leg.append('PWQMN '+csv_PWQMN.split("_")[-1][:-4])

    ax.grid()
    # sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
    ax.set_ylabel(p[0]+', mg/L')
    ax.set_xlabel('')
    # assign locator and formatter for the xaxis ticks.
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    
    
    ax.legend(leg)
    ax.set_title(p[0]+', mg/L')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.savefig(("/").join(csv_glni.split("/")[:3])+"/"+csv_glni.split("/")[2]+"_"+p[0].lower().replace(" ","_")+'.pdf')
```

```python
csv_glni =  './eastern_basin/lynn_river/GLNI/GLNI_lynn_creek_at_simcoe_ON02GC1100.csv'

csv_PWQMN_1 = './eastern_basin/lynn_river/PWQMN/PWQMN_lynn_river_16015900902.csv'
csv_PWQMN_2 = './eastern_basin/lynn_river/PWQMN/PWQMN_lynn_river_16015900302.csv'


for p in pairs:
    fig, ax = plt.subplots()
    df = pd.read_csv(csv_glni)
    leg = []
    df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
    try:
        if df[~df[p[0]].isnull()].shape[0]>0:
            leg.append('GLNI '+csv_glni.split("_")[-1][:-4])
            sns.scatterplot(x='SampleDateTime', y=p[0], data=df)
    except:
        pass
    try:
        for csv_PWQMN in [csv_PWQMN_1, csv_PWQMN_2]:
            df = pd.read_csv(csv_PWQMN)
            df['DATE'] = pd.to_datetime(df.DATE)
            if df[~df[p[1]].isnull()].shape[0]>0:
                sns.scatterplot(x='DATE', y=p[1], data=df)
                leg.append('PWQMN '+csv_PWQMN.split("_")[-1][:-4])
    except:
        pass
    ax.grid()
    # sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
    ax.set_ylabel(p[0]+', mg/L')
    ax.set_xlabel('')
    # assign locator and formatter for the xaxis ticks.
    ax.xaxis.set_major_locator(mdates.AutoDateLocator())
    
    
    ax.legend(leg)
    ax.set_title(p[0]+', mg/L')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.savefig(("/").join(csv_glni.split("/")[:3])+"/"+csv_glni.split("/")[2]+"_"+p[0].lower().replace(" ","_")+'.pdf') 
```

```python
csvs = glob.glob('./*/*/*/GLNI*.csv')
for river in csvs:
    river = ("/").join(river.split("/")[:3])+"/"
    csv_glni = glob.glob(river+'GLNI/*.csv')[0]
    csvs_PWQMN = glob.glob(river+'PWQMN/*.csv')


    for p in pairs:
        fig, ax = plt.subplots()

        leg = []

        try:
            df = pd.read_csv(csv_glni)
            df['SampleDateTime'] = pd.to_datetime(df.SampleDateTime)
            if df[~df[p[0]].isnull()].shape[0]>0:
                leg.append('GLNI '+csv_glni.split("_")[-1][:-4])
                sns.scatterplot(x='SampleDateTime', y=p[0], data=df)
        except:
            pass
        try:
            for csv_PWQMN in csvs_PWQMN:
                df = pd.read_csv(csv_PWQMN)
                df['DATE'] = pd.to_datetime(df.DATE)
                if df[~df[p[1]].isnull()].shape[0]>0:
                    sns.scatterplot(x='DATE', y=p[1], data=df)
                    leg.append('PWQMN '+csv_PWQMN.split("_")[-1][:-4])
        except:
            pass
        ax.grid()
        # sns.tsplot(df, time='Date', value='Value', unit='Unit', ax=ax)
        ax.set_ylabel(p[0]+', mg/L')
        ax.set_xlabel('')
        # assign locator and formatter for the xaxis ticks.
        ax.xaxis.set_major_locator(mdates.AutoDateLocator())


        ax.legend(leg)
        ax.set_title(river.split("/")[-2].title().replace("_", " ")+", "+ p[0]+', mg/L')
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
        plt.savefig(river+csv_glni.split("/")[2]+"_"+p[0].lower().replace(" ","_")+'.pdf') 
```

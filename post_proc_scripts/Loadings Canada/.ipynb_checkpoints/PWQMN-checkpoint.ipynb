```python
dfsn = []
for df in dfs:
    df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
    for c in df.columns:
        df = df.rename(index=str, columns={c:c.replace("_", " ")})
    df.rename({"STATION NO":"STATION", "LIMS_R": 'LIMS RCODE', "FIELD_NO":"FIELD NO", "FIELDNO":"FIELD NO"}, inplace=True)
    dfsn.append(df)
```

```python
pwqmn_rawdata = pd.concat(dfsn, ignore_index=True)
```

```python
river = 'Talfourd Creek'
station = 15000200202
elem = 'PHOSPHORUS,UNFILTERED TOTAL'
short_name = 'TP'
TP = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
# TP.columns = ['TP_'+str(col) for col in TP.columns]
TP.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
TP

elem = 'PHOSPHATE,FILTERED REACTIVE'
PO4 = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
PO4.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
PO4


elem = 'CHLORIDE,         UNFIL.REAC'
Cl = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
Cl.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
Cl

elem = 'SILICATES,UNFILTERED REACTIVE'
Si = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
Si.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
Si

elem = 'NITROGEN; TOTAL'
TN = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
TN.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
TN

elem = 'NITRITE,  UNFILTERED REACTIVE'
NO2 = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
NO2.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
NO2

elem = 'NITRATE,  UNFILTERED REACTIVE'
NO3 = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
NO3.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
NO3

elem = 'NITROGEN,TOT,KJELDAHL/UNF.REA'
NTK = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
NTK.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
NTK

elem = 'NITRATES TOTAL,   UNFIL.REAC'
NO3R = pwqmn_rawdata[(pwqmn_rawdata["PARM DESCRIPTION"]==elem) & ((pwqmn_rawdata["STATION"]==station) | (pwqmn_rawdata["STATION NO"]==station))][['DATE', 'RESULT', 'UNITS']]
NO3R.rename(columns={'UNITS':'UNITS '+elem, 'RESULT': elem}, index=str, inplace=True)
NO3R


dfs = [PO4, TN, NO3, NO2, NTK, NO3R, Si, Cl]
for df in dfs:
    TP = pd.merge(TP, df, on='DATE', how='outer')
TP.loc[TP['UNITS PHOSPHORUS,UNFILTERED TOTAL'] == 'MICROGRAM PER LITER', 'PHOSPHORUS,UNFILTERED TOTAL'] = TP.loc[TP['UNITS PHOSPHORUS,UNFILTERED TOTAL'] == 'MICROGRAM PER LITER', 'PHOSPHORUS,UNFILTERED TOTAL'] / 1000
TP.loc[TP['UNITS PHOSPHORUS,UNFILTERED TOTAL'] == 'MICROGRAM PER LITER', 'UNITS PHOSPHORUS,UNFILTERED TOTAL'] = 'MILLIGRAM PER LITER'

directory = 'st_clair_river/'+river.lower().replace(" ", "_")+'/'

if not os.path.exists(directory):
    os.makedirs(directory)

TP.to_csv(directory+"PWQMN_"+river.lower().replace(" ", "_")+"_"+str(station)+'.csv')

```

```python
station_dbln6 = pd.read_csv('../../measurements/Meteo/station_dbln6.csv', skiprows=0)
```

```python
station_dbln6['WSP'] = pd.to_numeric(station_dbln6['WSP'], errors='coerce')
station_dbln6['WD'] = pd.to_numeric(station_dbln6['WD'], errors='coerce')
station_dbln6['ATMP'] = pd.to_numeric(station_dbln6['ATMP'], errors='coerce')
station_dbln6['YY'] = pd.to_numeric(station_dbln6['YY'], errors='coerce')
station_dbln6['MM'] = pd.to_numeric(station_dbln6['MM'], errors='coerce')
station_dbln6['MM'] = pd.to_numeric(station_dbln6['MM'], errors='coerce')
station_dbln6['DD'] = pd.to_numeric(station_dbln6['DD'], errors='coerce')
# pd.to_datetime(station_dbln6[['MM','YY']])
```

```python
station_dbln6 = station_dbln6[station_dbln6.YY>2002]
```

```python
station_dbln6.WD = (90 - station_dbln6.WD) % 360
station_dbln6['U'] = -station_dbln6['WSP']*np.cos(np.deg2rad(station_dbln6.WD))
station_dbln6['V'] = -station_dbln6['WSP']*np.sin(np.deg2rad(station_dbln6.WD))
```

```python
station_dbln6 = station_dbln6.groupby(['YY','MM']).mean().reset_index()
```

```python
station_dbln6['U'], station_dbln6['V'] = [station_dbln6['U'], station_dbln6['V']]/np.linalg.norm([station_dbln6['U'], station_dbln6['V']],axis=0)
```

## Air temperature based on GLERL data

```python
years=np.arange(2003,2015)
```

```python
av_temp = pd.read_excel('../../measurements/Meteo/AirTemperature_OverBasin.xlsx', sheet_name='EriAve', skiprows=3)
av_temp = av_temp[55:67]
av_temp = av_temp.drop(columns=['Annual'])
```

```python
min_temp = pd.read_excel('../../measurements/Meteo/AirTemperature_OverBasin.xlsx', sheet_name='EriMin', skiprows=3)
min_temp = min_temp[55:67]
min_temp = min_temp.drop(columns=['Annual'])
```

```python
max_temp = pd.read_excel('../../measurements/Meteo/AirTemperature_OverBasin.xlsx', sheet_name='EriMax', skiprows=3)
max_temp = max_temp[55:67]
max_temp = max_temp.drop(columns=['Annual'])
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',       'Oct', 'Nov', 'Dec']
```

```python
for c in av_temp.columns:
    av_temp[c] = pd.to_numeric(av_temp[c], errors='coerce')
    min_temp[c] = pd.to_numeric(min_temp[c], errors='coerce')
    max_temp[c] = pd.to_numeric(max_temp[c], errors='coerce')
```

```python
min_temp[m].values
```

```python
fig, axes=plt.subplots(12,1, figsize=(15*0.8,13*1.5*0.8), sharex=True, dpi=100)
for i, m in enumerate(months):
    axes[i].plot(av_temp.Year, av_temp[m], c=sns.color_palette()[0])
    axes[i].fill_between(min_temp.Year.values, min_temp[m].values, max_temp[m].values, color=sns.color_palette()[0], alpha=0.1)
#     axes[i].plot(max_temp.Year, max_temp[m], c=sns.color_palette()[0], lw=0.5)
#     axes[i].set_ylim(-10,25)
    axes[i].set_ylim(np.floor(axes[i].get_ylim()[0]), np.ceil(axes[i].get_ylim()[1]))
    axes[i].set_xlim(2003,2014)
    axes[i].legend(loc=1)
#     axes[i].text(0.99, 0.9, m, horizontalalignment='center', verticalalignment='center', transform=axes[i].transAxes)
    axes[i].grid(lw=0.2)
#     plt.legend(loc=(1.021,0))
axes[-1].text(-0.06, 6, 'Air Temperature, C', fontsize=14, horizontalalignment='center', verticalalignment='center', transform=axes[-1].transAxes, rotation=90)

# fig.savefig('Mass Balance/GLENDA/All Stations/Temperature_Lake_Monthly_GLERL.pdf')
```

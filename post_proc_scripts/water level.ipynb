```python
bloom_sever = pd.read_csv('../measurements/bloom severity noaa.txt')
bloom_sever
```

```python
from matplotlib import rcParams
rcParams.update({'axes.titlesize': 'large', 'xtick.labelsize':'x-small', 'axes.labelweight': 'bold', 'ytick.labelsize':'x-small'})

fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(df.year.unique()[-15:], df.groupby('year').max().Erie[-15:] - df[df.month==1].Erie[-15:].values, lw=5, color=sns.color_palette("deep")[0])
ax1.set_ylim(-0.8,0.8)

# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('$\Delta$ Water Level (MAX - JAN), m ', color=sns.color_palette("deep")[0], fontsize=20)
ax1.tick_params('y', colors=sns.color_palette("deep")[0])
ax1.set_axisbelow(True)
ax1.yaxis.grid(color='black', linestyle='dashed', alpha=0.1)
ax1.xaxis.grid(color='black', linestyle='dashed', alpha=0.1)

ax2 = ax1.twinx()
ax2.bar(bloom_sever.Year, bloom_sever.Severity, color=sns.color_palette("deep")[1], edgecolor = "k", linewidth=0.7)
ax2.set_ylim(0,16)
ax2.set_ylabel('Bloom Severity Index by NOAA', color=sns.color_palette("deep")[1], fontsize=20)
ax2.tick_params('y', colors='g')
ax2.set_axisbelow(True)
ax2.yaxis.grid(color='black', linestyle='dashed', alpha=0.1)
ax2.xaxis.grid(color='black', linestyle='dashed', alpha=0.1)

ax2.spines['top'].set_color('black')
ax2.spines['bottom'].set_color('black')
ax2.spines['right'].set_color(sns.color_palette("deep")[1])
ax2.spines['left'].set_color(sns.color_palette("deep")[0])


# ax1.text(2014.9, 0.64, 'r = 0.7\n p = 0.003', fontsize=20, bbox={'facecolor':'white', 'alpha':1, 'pad':7})

fig.tight_layout()
plt.savefig('wl_vs_blooms.pdf')
plt.show()
```

```python
from datetime import datetime
mdf['Date'] = mdf.apply(lambda x:datetime.strptime("{0} {1} {2} 00:00:00".format(int(x['Year']),int(x['Month']), int(x['Day'])), "%Y %m %d %H:%M:%S"),axis=1)
ddf['Date'] = ddf.apply(lambda x:datetime.strptime("{0} {1} {2} 00:00:00".format(int(x['Year']),int(x['Month']), int(x['Day'])), "%Y %m %d %H:%M:%S"),axis=1)
```

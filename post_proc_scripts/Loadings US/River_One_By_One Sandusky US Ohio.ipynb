### Discharge

```python
# Finde station here:
# https://www.hydrocloud.org
site = '04198000'
start = '1996-01-01'
end = '2018-11-01'
response = hf.get_nwis(site, 'dv',start, end)
# response.get_data()
portage_Q = hf.extract_nwis_df(response.json())
```

```python
portage_Q['Flow, CFS']=portage_Q['USGS:04198000:00060:00003']
```

```python
portage_Q= portage_Q.reset_index()
portage_Q['Date']  = portage_Q['datetime']
portage_Q['Date'] =pd.to_datetime(portage_Q['Date'], errors='coerce')
portage_Q = portage_Q[['Date', 'Flow, CFS']]
portage_Q['Source'] = 'USGS-04198000'
```

```python
df_Q_heideleberg = pd.read_csv('/Users/imarkelo/git/Great-Lakes-Data/post_proc_scripts/Heidelberg/proc_data/Flow_Sandusky_River_data.csv')
df_Q_heideleberg['Date'] =pd.to_datetime(df_Q_heideleberg['Date'], errors='coerce')
df_Q_heideleberg['Source'] = 'HTLP (NCWQR)'
```

```python
portage_Q = portage_Q.append(df_Q_heideleberg)
```

```python
portage_Q[portage_Q.Source=='USGS-04198000'][['Date', 'Flow, CFS']].dropna().to_csv('One_by_one/Western/US/Sandusky R/Flow_Sandusky_R_CFS.csv', index=False)
```

```python
portage_Q['Discharge, cms'] = 0.028*portage_Q['Flow, CFS']
```

```python
river = 'Sandusky_R'
```

```python
fig,ax=plt.subplots(1,1,figsize=(6,4), dpi=100)
sns.lineplot(x='Date', y='Discharge, cms', data=portage_Q, hue='Source', hue_order=['USGS-04198000', 'HTLP (NCWQR)'])
plt.legend(['(10) - USGS-04198000', '(9) - HTLP (NCWQR)'], loc=2, fontsize='small')
plt.ylim(0,1000)
plt.xlim(datetime.datetime(1996,1,1), datetime.datetime(2018,1,1))
plt.grid(alpha=0.5)
plt.ylabel("Discharge, m$^3$/s")
plt.xlabel("Year")
plt.savefig('One_by_one/Western/US/'+river+'/'+river+'_Q.pdf')
plt.savefig('/Users/imarkelo/Google Drive/GDocuments/Lake Erie/Draft/Rivers_tables/img/Western/US/'+river+'/'+river+'_Q.pdf')
```

### EGRET Results

```python
river = 'Sandusky_R'
EGRET_Daily = pd.read_csv('One_by_one/Western/US/'+river+'/EGRET Result/TP/'+river+'iver_TP_Daily.csv')
EGRET_Sample = pd.read_csv('One_by_one/Western/US/'+river+'/EGRET Result/TP/'+river+'iver_TP_Sample.csv')
```

```python
EGRET_Sample['Fx_Hat'] = EGRET_Sample['Q']*60*60*24*1000 * EGRET_Sample['ConcHat']/1000/1000
EGRET_Sample['Fx_C'] = EGRET_Sample['Q']*60*60*24*1000 * EGRET_Sample['ConcAve']/1000/1000
```

```python
fig,ax=plt.subplots(1,1,figsize=(6,4), dpi=100)
sns.scatterplot(y='Fx_C', 
                x='Fx_Hat',
                data=EGRET_Sample,
                s=15,
                linewidth=0.3, ax=ax,
               color=sns.color_palette("Blues")[4])
# plt.legend("")
ax.set_xscale('log')
ax.set_yscale('log')
plt.plot([-100,100000],[-100,100000], 'k', lw=0.5)
plt.xlim(0,10_000)
plt.ylim(0,10_000)
# plt.xlim(datetime.datetime(1996,1,1), datetime.datetime(2018,1,1))
plt.grid(alpha=0.5)
plt.xlabel("Estimated Phosphorus Loads, kg/day")
plt.ylabel("Observed Phosphorus Loads, kg/day")
# plt.xlabel("")
# plt.savefig('One_by_one/Western/US/Portage R/Portage Q.pdf')
plt.tight_layout()
# plt.savefig('/Users/imarkelo/Google Drive/GDocuments/Lake Erie/Draft/Rivers_tables/img/Western/US/'+river+'/'+river+'_FvsF.pdf')
```

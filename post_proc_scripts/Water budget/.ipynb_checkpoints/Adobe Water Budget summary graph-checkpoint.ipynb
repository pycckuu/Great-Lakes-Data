```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

sns.set_style("whitegrid")
sns.set_style("ticks")
from datetime import datetime
str(datetime.now())

import scipy as sp
```

```python
table = pd.read_excel('/Users/imarkelo/Google Drive/GDocuments/Lake Erie/Draft/Water Budget/Water Budget_gauged.xlsx', sheet_name='Total Annual 0.75 ERA5')
```

```python
table
```

```python
st_clair = table.iloc[0][1:-1]/table.iloc[0][-1]*100 - 100
```

```python
f, axes = plt.subplots(1,1, figsize=(1,0.65), dpi=600, sharex=True)

axes.axhline(0, c='k', ls='--', lw=0.3)
axes.plot(st_clair.index.values, st_clair.values, lw=1, c='k', label='St.Clair')
axes.set_xlabel('')
# axes.grid(lw=0.2)
# axes.set_ylim(-15,15)
# axes.set_xlim(3,17)
# axes.set_ylabel('%')
for axis in ['top','bottom','left','right']:
    axes.spines[axis].set_linewidth(0.3)
# axes.text(1.2,0.35, str(table.iloc[0][-1]).split('.')[0], horizontalalignment='center', verticalalignment='center', transform=axes.transAxes)
axes.yaxis.set_tick_params(width=0.3)
axes.set_xlabel('')
# plt.xticks(rotation='45')
# axes.legend(loc=2, frameon=1, framealpha=1)
axes.get_xaxis().set_ticklabels([])
# for label in (axes.get_xticklabels() + axes.get_yticklabels()):
#     label.set_fontname('Arial')
#     label.set_fontsize(5)
# axes.get_xaxis().set_ticklabels([])
axes.get_xaxis().set_ticks([])
axes.get_yaxis().set_ticks([-10,10])
axes.get_yaxis().set_ticklabels(['-10%', '10%'])
sns.despine(offset=1, trim=True,bottom=True);
plt.tight_layout()
f.savefig('img/clair_river.pdf')
```

```python
f, axes = plt.subplots(2,2, figsize=(1*2,1*2), dpi=600, sharex=True, sharey=True)

labels = ['US', 'CA', 'P', 'E']
colors = [0,3,1,2]

for i,ax1 in enumerate(axes):
    for j, ax in enumerate(ax1):
        ax.axhline(0, c='k', ls='--', lw=0.3)
        ax.plot(table.iloc[(i+1)*i+j+1][1:-1]/table.iloc[(i+1)*i+j+1][-1]*100 - 100, lw=1, label='St.Clair', c=sns.color_palette()[colors[(i+1)*i+j]])
        ax.set_xlabel('')
        # axes.grid(lw=0.2)
        # axes.set_ylim(-15,15)
        # axes.set_xlim(3,17)
        # axes.set_ylabel('%')
        ax.yaxis.set_tick_params(width=0.3)
        for axis in ['top','bottom','left','right']:
            ax.spines[axis].set_linewidth(0.3)
        ax.text(0.5,1.1, labels[(i+1)*i+j], horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
        ax.text(1.2,0.38, str(round(table.iloc[(i+1)*i+j+1][-1])).split('.')[0], horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, weight='bold')
        ax.set_xlabel('')
        # plt.xticks(rotation='45')
        # axes.legend(loc=2, frameon=1, framealpha=1)
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticklabels(['0', '50%'])

plt.subplots_adjust(hspace=.7, wspace=.9, left=0.2, right=0.88)
sns.despine(offset=1, trim=True,bottom=True);
# plt.tight_layout()
f.savefig('img/clair_lake.pdf')

```
